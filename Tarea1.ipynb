# Ejemplo de Regresión lineal


```python
# Importaciones necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
import warnings
warnings.filterwarnings('ignore')

# Cargar el conjunto de datos
loan_data = pd.read_csv('C:/Users/diego/OneDrive/Escritorio/loan_data_2015.csv')


# Preparar los datos para la regresión
X_loan_amnt = loan_data[['loan_amnt']]
y_installment = loan_data['installment']

# Realizar la regresión lineal
lr_installment = LinearRegression()
lr_installment.fit(X_loan_amnt, y_installment)
loan_data['predicted_installment'] = lr_installment.predict(X_loan_amnt)

# Visualizar los datos y la línea de regresión
plt.figure(figsize=(10, 6))
sns.scatterplot(x=loan_data['loan_amnt'], y=loan_data['installment'], alpha=0.5)
sns.lineplot(x=loan_data['loan_amnt'], y=loan_data['predicted_installment'], color='red')
plt.title('Regresión Lineal: loan_amnt vs installment')
plt.xlabel('loan_amnt (Monto del Préstamo)')
plt.ylabel('installment (Pago Mensual)')
plt.xlim(left=loan_data['loan_amnt'].quantile(0.01), right=loan_data['loan_amnt'].quantile(0.99))  # Limitando para visualizar mejor
plt.show()
```
